# Exibindo Gráficos no Python

### Base

```python
vendas_meses = [2907,1402,1666,2052,1007,2673,1996,2296,1080,2934,1505,2442]
meses = ['jan',	'fev',	'mar',	'abr',	'mai',	'jun',	'jul',	'ago',	'set',	'out',	'nov',	'dez']
```

### Importações

```python
import matplotlib.pyplot as plt
```

### Plotando Gráfico

```python
plt.plot(vendas_meses)
plt.show()
```

* #### Adicionando Labels

```python
plt.plot(vendas_meses)
plt.ylabel('Vendas')
plt.xlabel('Meses')
plt.show()
```

* #### Adicionando Meses

```python
plt.plot(meses, vendas_meses)
plt.ylabel('Vendas')
plt.xlabel('Meses')
plt.show()
```

* #### Escala Eixo Y

```python
plt.plot(meses, vendas_meses)
plt.ylabel('Vendas')
plt.xlabel('Meses')
plt.axis([0,12,1000,max(vendas_meses)+300])
plt.show()
```
