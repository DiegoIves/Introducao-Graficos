# Exibindo Gráficos no Python

### Importações

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Base

```python
vendas = np.random.randint(1000,3000,50)
meses = np.arange(1,51)
```

### Plotando Gráfico

* #### Grafico de Linha

```python
plt.plot(meses,vendas)
plt.axis([0,50,0,max(vendas)+250])
plt.ylabel("Vendas")
plt.xlabel("Meses")
plt.show()
```

* #### Gráfico de Pontos

```python
#marcadores azuis
plt.plot(meses,vendas,'bo')
plt.axis([0,50,0,max(vendas)+250])
plt.ylabel("Vendas")
plt.xlabel("Meses")
plt.show()
#tracejado amarelo
plt.plot(meses,vendas,'c--')
plt.axis([0,50,0,max(vendas)+250])
plt.ylabel("Vendas")
plt.xlabel("Meses")
plt.show()
#quadrado verde
plt.plot(meses,vendas,'ms')
plt.axis([0,50,0,max(vendas)+250])
plt.ylabel("Vendas")
plt.xlabel("Meses")
plt.show()
#triangulo vermelho
plt.plot(meses,vendas,'r^')
plt.axis([0,50,0,max(vendas)+250])
plt.ylabel("Vendas")
plt.xlabel("Meses")
plt.show()
```

* #### Dispersão

```python
plt.scatter(meses,vendas)
plt.axis([0,50,0,max(vendas)+250])
plt.ylabel("Vendas")
plt.xlabel("Meses")
plt.show()
```

* #### Barras

```python
plt.bar(meses,vendas,)
plt.axis([0,51,0,max(vendas)+250])
plt.ylabel("Vendas")
plt.xlabel("Meses")
plt.show()
```

* #### Multiplos Gráficos

```python
#Definir figura
plt.figure(figsize=(16,4))
#plotar graficos linhas, graficos, ordem
plt.subplot(1,3,1)
plt.plot(meses,vendas,'ms')
plt.subplot(1,3,2)
plt.bar(meses,vendas,)
plt.subplot(1,3,3),
plt.scatter(meses,vendas)
```
